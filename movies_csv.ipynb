```python
movies[movies.title.str.startswith('Once')]
```

same thing using user defined function calling:- 

```python
import re 

def check(x):
    m=re.search('^once',x,re.I)
    if m:
        return True
    else:
        return False
    
movies[movies.title.apply(check)]          
               
```
